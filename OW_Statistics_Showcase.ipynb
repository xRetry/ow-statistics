```
# Getting the tool from Github and importing it into the notepad
!git clone 'https://github.com/xRetry/ow_statistics.git'
import ow_statistics.ow_statistics as ow
```

```
# Loading the data of all matches
# The file(s) has to uploded before it can be used. 
# NOTE: This usually takes some time. Look button left to see when it is done.
# Can also take multiple files as list: ['file1.json', 'file2.json', ...]
stats = ow.DataFrame('desolation_only.json')
```

```
# To see stats, a match has to be selected first.
# This can be done by the outfit tag or the zone ID
stats.set_match(outfit_tag='RE4')
```

```
# After that, various plots are available.
# Every plot has some options to put in the brackets.
# If they are empty, the default options are used.
stats.plot_timeline_facility()
```

```
# In this case specific factions are selected.
# Not all plots allow this though. For example Timelines don't.
stats.plot_players_revives('NC', 'TR')
```

```
# For some you can specify wheter or not to include revives.
stats.plot_timeline_kdr(with_revives=False)
```
